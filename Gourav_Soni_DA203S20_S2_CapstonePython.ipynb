### Task 7

#### How are popularities of movies related with the movie budgets? Are they correlated or totally uncorrelated with each other? write the interpretation of your analysis.

```python
# Calculate the correlation between popularity and budget
correlation = movie_data['popularity'].corr(movie_data['budget'])

# Display the correlation coefficient
print(f"Correlation between popularity and budget: {correlation}")
```

### Interpretation:

#### If the correlation coefficient is close to 1, it suggests a strong positive correlation, meaning as the movie budget increases, the popularity tends to increase.
#### If the correlation coefficient is close to -1, it suggests a strong negative correlation, meaning as the movie budget increases, the popularity tends to decrease.
#### If the correlation coefficient is close to 0, it suggests a weak or no correlation, meaning there is no clear linear relationship between movie budgets and popularity.
#### Keep in mind that correlation does not imply causation, and other factors could influence the relationship between movie budgets and popularity. It's always good to complement statistical analysis with domain knowledge and other forms of exploration

### Task 10

#### Sort the data in descending order based on revenue and filter the top 500 movies. Find measures of central tendency for the following columns using the filter data.
#### budget, revenue and runtime

#### perform outerlier analysis for above 3 col. using box plot chart.

```python
# Sort the data in descending order based on revenue
movie_data_sorted = movie_data.sort_values(by='revenue', ascending=False)

# Filter the top 500 movies
top_500_movies = movie_data_sorted.head(500)

# Display measures of central tendency for 'budget', 'revenue', and 'runtime'
measures_of_central_tendency = top_500_movies[['budget', 'revenue', 'runtime']].describe()
print("Measures of central tendency for budget, revenue, and runtime:")
print(measures_of_central_tendency)

# Outlier analysis using box plot
plt.figure(figsize=(12, 6))

# Box plot for 'budget'
plt.subplot(1, 3, 1)
sb.boxplot(y=top_500_movies['budget'])
plt.title('Budget')

# Box plot for 'revenue'
plt.subplot(1, 3, 2)
sb.boxplot(y=top_500_movies['revenue'])
plt.title('Revenue')

# Box plot for 'runtime'
plt.subplot(1, 3, 3)
sb.boxplot(y=top_500_movies['runtime'])
plt.title('Runtime')

plt.tight_layout()
plt.show()
```

### Task 11

#### Identify and display the names of the movies along with their runtimes for those movies that have above average runtime using the data from the previous task.

```python
# Assuming 'top_500_movies' is the DataFrame from the previous task

# Calculate the average runtime
average_runtime = top_500_movies['runtime'].mean()

# Filter movies with above-average runtime
above_average_runtime_movies = top_500_movies[top_500_movies['runtime'] > average_runtime]

# Display names of movies along with their runtimes
movie_names_runtimes = above_average_runtime_movies[['title', 'runtime']]
print("Movies with above-average runtime:")
movie_names_runtimes
```
